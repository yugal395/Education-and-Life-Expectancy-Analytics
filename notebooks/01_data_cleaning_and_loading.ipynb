```python
leboth_clean=leboth.copy()

#1) Rename columns to easier to easier name(no spaces, no paranthese)
leboth_clean=leboth_clean.rename(columns={"Country or Area":"Country","Year(s)":"Year","Variant":"Variant","Value":"Value"})

#2) converting year to a number
leboth_clean["Year"]=pd.to_numeric(leboth_clean["Year"],errors="coerce")

#3) convert life expectancy to a nummber(float)
leboth_clean["Value"]=pd.to_numeric(leboth_clean["Value"],errors="coerce")

#quickcheck
leboth_clean.head()
leboth_clean.describe()
```

```python
leboth_clean["Variant"].value_counts().head(10)
#that means the whole dataset has only medium variant but we need to confirm it
leboth_clean["Variant"].nunique(), leboth_clean["Variant"].unique()
```

```python
#as its confirmed now, we can drop the variant column as it has only one value
leboth_clean=leboth_clean.drop(columns=["Variant"])
```

```python
leboth=leboth_clean.copy()
```

```python
leboth.info()
```

```python
print("Missing values per column")
print(leboth.isna().sum())

print("duplicate country year rows")
print(leboth.duplicated(subset=["Country","Year"]).sum())
```

```python
leboth=leboth.dropna(subset=["Value"]).copy()
print("After dropping missing values",leboth.shape)  # we drop the rows missing values(none) and copied it to our main leboth df
```

```python
# now we can use the data for only years like 2025 so that we dont go into prediction, i mean thats what we want to do right and not just plot graphs for life expectancy over years vs education
leboth=leboth[leboth["Year"]<=2025].copy() # this stores only the dataframe column which has year less than equal to 2025
leboth["Year"].min(), leboth["Year"].max()
```

```python
leboth.shape
```

```python
leboth["Country"]=leboth["Country"].astype(str).str.strip()
#.astype(str)
#→ makes sure everything in Country is text

#.str.strip()
#→ removes extra spaces at the start or end of country names

leboth=leboth.rename(columns={"Value": "LifeExp_Both"})
```
