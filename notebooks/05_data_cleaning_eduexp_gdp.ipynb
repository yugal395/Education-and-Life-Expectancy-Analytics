```python
import pandas as pd


exp_gdp_raw=pd.read_csv("../data/Expenditure_by_GDP.csv")
exp_gdp_raw.head()
```

```python
print(exp_gdp_raw.shape)
print(exp_gdp_raw.columns)
exp_gdp_raw.info()
```

```python
print(exp_gdp_raw["Sex"].value_counts(dropna=False))
print(exp_gdp_raw["Age group"].value_counts(dropna=False))
print(exp_gdp_raw["Units of measurement"].value_counts(dropna=False))
```

```python

exp_gdp = exp_gdp_raw.copy()
exp_gdp = exp_gdp.rename(columns={"Reference Area": "Country","Time Period":"Year","Observation Value":"Value"})

exp_gdp.head()
```

```python
# filter to all genders (only if this column exists)
exp_gdp = exp_gdp[exp_gdp["Sex"] == "All genders"].copy()

# drop columns that are no longer useful
exp_gdp = exp_gdp.drop(columns=["Sex", "Age group", "Units of measurement"])

# convert Year and Value to numeric
exp_gdp["Year"] = pd.to_numeric(exp_gdp["Year"], errors="coerce")
exp_gdp["Value"] = pd.to_numeric(exp_gdp["Value"], errors="coerce")

# drop missing essentials
exp_gdp = exp_gdp.dropna(subset=["Country", "Year", "Value"]).copy()
exp_gdp["Year"] = exp_gdp["Year"].astype(int)

# clean country names
exp_gdp["Country"] = exp_gdp["Country"].astype(str).str.strip()

# rename Value to meaningful name
exp_gdp = exp_gdp.rename(columns={"Value": "EduExp_GDP"})
```

```python
exp_gdp.head()
print(exp_gdp.shape)
print("Duplicate Country-Year:",
      exp_gdp.duplicated(subset=["Country","Year"]).sum())
```

```python
exp_gdp.to_csv("../data/cleaned_education_expenditure_gdp.csv", index=False)
```
