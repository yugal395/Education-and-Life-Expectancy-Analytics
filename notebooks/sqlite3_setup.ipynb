```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../output/education_life_expectancy.db") #if a db doesnt exist, it will be created
print("Connected to sqlite3 database.")
```

```python
life_both=pd.read_csv("../output/life_expectancy_both_clean.csv")
life_both.head()
```

```python
life_both.to_sql("life_both", conn, if_exists="replace", index=False)
print("DataFrame written to sqlite3 database.")
#equivalent to python .import life_expectancy_both_clean.csv life_both
```

```python
pd.read_sql_query("SELECT * FROM life_both LIMIT 5;", conn)
```

```python
primary   = pd.read_csv("../output/primary_clean.csv")
secondary = pd.read_csv("../output/secondary_clean.csv")
tertiary  = pd.read_csv("../output/tertiary_clean.csv")
exp_gdp   = pd.read_csv("../output/cleaned_education_expenditure_gdp.csv")
exp_total = pd.read_csv("../output/exp_total_clean.csv")
```

```python
primary.to_sql("primary", conn, if_exists="replace", index=False)
secondary.to_sql("secondary", conn, if_exists="replace", index=False)
tertiary.to_sql("tertiary", conn, if_exists="replace", index=False)
exp_gdp.to_sql("exp_gdp", conn, if_exists="replace", index=False)
exp_total.to_sql("exp_total", conn, if_exists="replace", index=False)

print("All tables written to SQLite")
```
