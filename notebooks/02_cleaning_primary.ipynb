```python
import pandas as pd


#now we gonna clean the primary education dataset
primary_raw=pd.read_csv('../data/primary.csv')
primary_raw.head()
```

```python
print(primary_raw.shape)
print(primary_raw.columns)
primary_raw.info()
```

```python
primary=primary_raw.copy()
# rename columns so that we have same columns so we can join them to create a combined dataset
primary=primary.rename(columns={"Reference Area": "Country","Time Period":"Year","Observation Value":"Value"})

primary.head()
```

```python
#converting to numbers if they are strings or something else
primary["Year"]=pd.to_numeric(primary["Year"],errors="coerce")
primary["Value"]=pd.to_numeric(primary["Value"],errors="coerce")

primary["Units of measurement"].unique() # age group is also same ,so we can drop them
primary.head()
```

```python
primary_raw["Sex"].value_counts(dropna=False)
```

```python
primary=primary[primary["Sex"]=="All genders"].copy()
primary.head()
```

```python
primary=primary.drop(columns=["Sex","Age group","Units of measurement"])
primary.head()
primary.shape
```

```python
primary=primary.dropna(subset=["Country","Year","Value"]).copy()
primary.shape
```

```python
primary["Country"]=primary["Country"].astype(str).str.strip() # stripping the extra space in front or bebhidn of country name
```

```python
primary=primary.rename(columns={"Value":"Enroll_Primary"})
primary.shape
```

doesnt change anything as such, pretty clean data then

```python
print(primary["Year"].min(),primary["Year"].max())

print("Duplicate Country-Year pairs:",primary.duplicated(subset=["Country","Year"]).sum())
```

```python
primary.to_csv("../output/primary_clean.csv", index=False)
```
