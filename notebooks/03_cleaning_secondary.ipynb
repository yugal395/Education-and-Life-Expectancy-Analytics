```python
import pandas as pd
secondary_raw=pd.read_csv("../data/secondary.csv")
secondary_raw.head()
```

```python
print(secondary_raw.shape)
print(secondary_raw.columns)
secondary_raw.info()
```

```python
secondary = secondary_raw.copy()
secondary=secondary.rename(columns={"Reference Area": "Country","Time Period":"Year","Observation Value":"Value"})
secondary.head()
```

```python


# filter to all genders
secondary = secondary[secondary["Sex"] == "All genders"].copy()  

# drop columns that are no longer needed
secondary = secondary.drop(columns=["Sex", "Age group", "Units of measurement"])

# convert types
secondary["Year"] = pd.to_numeric(secondary["Year"], errors="coerce")
secondary["Value"] = pd.to_numeric(secondary["Value"], errors="coerce")

# drop missing
secondary = secondary.dropna(subset=["Country", "Year", "Value"]).copy()
secondary["Year"] = secondary["Year"].astype(int)

# clean country text
secondary["Country"] = secondary["Country"].astype(str).str.strip()

# rename value
secondary = secondary.rename(columns={"Value": "Enroll_Secondary"})
```

```python
secondary.head()
```

```python
#ceheckin  the max and min
print(secondary["Year"].min(),secondary["Year"].max())
print("Duplicate Country-Year pairs:",secondary.duplicated(subset=["Country","Year"]).sum())
```

```python
secondary.to_csv("../output/secondary_clean.csv", index=False)
```
