```python
import pandas as pd
# load raw data
exp_total_raw = pd.read_csv("../data/Expenditure_by_total_exp.csv")

# start cleaning
exp_total = exp_total_raw.copy()

# filter to all genders
exp_total = exp_total[exp_total["Sex"] == "All genders"].copy()

exp_total = exp_total_raw.rename(columns={"Reference Area": "Country","Time Period":"Year","Observation Value":"Value"})

exp_total.head()
# drop columns that are no longer useful
exp_total = exp_total.drop(columns=["Sex", "Age group", "Units of measurement"])

# convert Year and Value to numeric
exp_total["Year"] = pd.to_numeric(exp_total["Year"], errors="coerce")
exp_total["Value"] = pd.to_numeric(exp_total["Value"], errors="coerce")

# drop missing essentials
exp_total = exp_total.dropna(subset=["Country", "Year", "Value"]).copy()
exp_total["Year"] = exp_total["Year"].astype(int)

# clean country names
exp_total["Country"] = exp_total["Country"].astype(str).str.strip()

# rename Value to meaningful name
exp_total = exp_total.rename(columns={"Value": "EduExp_Total"})
```

```python
exp_total.head()
print(exp_total.shape)
print("Duplicate Country-Year:",
      exp_total.duplicated(subset=["Country","Year"]).sum())
```

```python
exp_total.to_csv("../output/exp_total_clean.csv", index=False)
print("Saved output/exp_total_clean.csv")
```
